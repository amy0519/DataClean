# 讀取原始資料並顯示所有欄位

## 篩選資料

```python
df2 = df[(df['交易標的']!='土地') & (df['交易標的']!='車位')]
df2
```

## 將索引重設定 

```python
df2 = df2.reset_index(drop='true')
df2
```

```python
df = df2.rename(columns = {'土地移轉總面積平方公尺':'土地移轉總面積(坪)','建物移轉總面積平方公尺':'建物移轉總面積(坪)','車位移轉總面積平方公尺':'車位移轉總面積(坪)','單價每平方公尺':'每坪單價'})
df
```

```python
df.iloc[:,3] = df.iloc[:,3] *3.30579
df.iloc[:,15] = df.iloc[:,15] *3.30579
df.iloc[:,24] = df.iloc[:,24] *3.30579
df.iloc[:,22] = df.iloc[:,22] *3.30579

df.head()
```

```python
df.loc[:,'總價元'] = df.loc[:,'總價元']/10000 #改成以萬為單位，以利後續統計圖表顯示
df
```

```python
df.groupby('交易標的').mean()
```

```python
print(df.loc[0,'交易年月日'])
print(type(df.loc[0,'交易年月日']))
```

```python
for i in range(len(df)):
    df.loc[i,'交易年'] = round(df.loc[i,'交易年月日']/10000)
    
df
```

## 取得西元年 

```python
from datetime import datetime
now = datetime.now()
now.year
#print(now.month)
```

```python
print(df.loc[0,'建築完成年月'])
print(type(df.loc[0,'建築完成年月']))
```

```python
for i in range(len(df)):
    df.loc[i,'建築完成年月'] = round(df.loc[i,'建築完成年月']/10000)
    
df
```

```python
for i in range(len(df)):
    df.loc[i,'屋齡'] = now.year - df.loc[i,'建築完成年月']-1911 #將西元-建築完成年月後轉成民國年

df
```

## 相關性分析

```python
corr = df[['總價元','土地移轉總面積(坪)','建物移轉總面積(坪)','建物現況格局-房','建物現況格局-廳','建物現況格局-衛','車位移轉總面積(坪)','車位總價元','屋齡','建築完成年月','交易年']].corr()
corr
```

```python
import seaborn as sns
plt.rcParams['font.family']='Microsoft JhengHei' #顯示中文
plt.figure(figsize=(10,10))
sns.heatmap(corr, square=True, annot=True)
sns.plt.show()
```
